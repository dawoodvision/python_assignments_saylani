3. Write a program in which use all the operators we can use in Python

```python
a = 10
b = 3

# Arithmetic Operators
print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)
print("Modulus:", a % b)

# Comparison Operators
print("Is a greater than b:", a > b)
print("Is a less than b:", a < b)
print("Is a equal to b:", a == b)
print("Is a not equal to b:", a != b)

# Assignment Operators
x = 5
x += 2
print("After adding 2:", x)
x -= 2
print("After subtracting 2:", x)
x *= 2
print("After multiplying by 2:", x)

# Logical Operators
print("Is a greater than 5 and b less than 5:", a > 5 and b < 5)
print("Is a greater than 5 or b greater than 5:", a > 5 or b > 5)
print("Is a not greater than 5:", not a > 5)
```

4. Completes the following steps of small task:

```python
english = 85
islamiat = 90
maths = 78

total_marks = 300

percentage = (english + islamiat + maths) / total_marks * 100

print("Percentage:", percentage, "%")
```
